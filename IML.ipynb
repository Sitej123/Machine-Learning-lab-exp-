```python
x=np.array([[1,1,1],[1,1,2],[1,2,2]])
y=np.array([[6],[8],[9]])
mult=np.multiply(x,y)
# print(mult)
x_t=np.transpose(x)
x_tx=np.matmul(x_t,x)
inv=np.linalg.inv(x_tx)
x_ty=np.matmul(x_t,y)
y_pred=np.matmul(inv,x_ty)
print(y_pred)
print(f"the regression equation is y={int(y_pred[0][0])} + {int(y_pred[1][0])} x1 +{int(y_pred[2][0])} x2 ")
```
