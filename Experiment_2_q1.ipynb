```python
n = len(y)
p = x.shape[1] - 1 # Number of features, excluding the intercept

# Calculate MAE
mae = np.mean(np.abs(y - y_pred))

# Calculate MSE
mse = np.mean((y - y_pred)**2)

# Calculate RMSE
rmse = np.sqrt(mse)

# Calculate R-squared (R2)
ss_total = np.sum((y - np.mean(y))**2)
ss_residual = np.sum((y - y_pred)**2)

# Handle the case where ss_total might be zero to prevent division by zero
if ss_total == 0:
    r2 = 1.0 if ss_residual == 0 else np.nan
else:
    r2 = 1 - (ss_residual / ss_total)

# Calculate Adjusted R-squared with a check for division by zero
if (n - p - 1) > 0:
    adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)
else:
    adjusted_r2 = np.nan # Cannot calculate adjusted R2 if denominator is zero or negative

print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R-squared (R2): {r2:.4f}")
print(f"Adjusted R-squared (Adjusted R2): {adjusted_r2:.4f}")
```
