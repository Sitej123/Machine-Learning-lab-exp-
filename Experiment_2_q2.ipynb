```python
import pandas as pd

x = [1, 2, 3]
y1 = [2, 4, 6]
y2 = [3, 5, 7]

df = pd.DataFrame({'x': x, 'y1': y1, 'y2': y2})

print("DataFrame created successfully:")
print(df.head())
```

```python
from sklearn.linear_model import LinearRegression

# Reshape the independent variable x into a 2D array
X = df['x'].values.reshape(-1, 1)

# Create and fit a LinearRegression model for y1
model_y1 = LinearRegression()
model_y1.fit(X, df['y1'])

# Create and fit a LinearRegression model for y2
model_y2 = LinearRegression()
model_y2.fit(X, df['y2'])

print("Linear Regression models for y1 and y2 fitted successfully.")
```

```python
y1_coefficient = model_y1.coef_[0]
y1_intercept = model_y1.intercept_

print(f"Regression Equation for y1:")
print(f"Coefficient (slope) for y1: {y1_coefficient:.2f}")
print(f"Intercept for y1: {y1_intercept:.2f}")
print(f"Equation: y1 = {y1_coefficient:.2f} * x + {y1_intercept:.2f}\n")

y2_coefficient = model_y2.coef_[0]
y2_intercept = model_y2.intercept_

print(f"Regression Equation for y2:")
print(f"Coefficient (slope) for y2: {y2_coefficient:.2f}")
print(f"Intercept for y2: {y2_intercept:.2f}")
print(f"Equation: y2 = {y2_coefficient:.2f} * x + {y2_intercept:.2f}")
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# 1. Make predictions for 'y1'
y1_pred = model_y1.predict(X)

# 2. Make predictions for 'y2'
y2_pred = model_y2.predict(X)

# 3. Calculate residuals for 'y1'
y1_residuals = df['y1'] - y1_pred

# 4. Calculate residuals for 'y2'
y2_residuals = df['y2'] - y2_pred

print("--- Error Metrics for y1 ---")
# 6. Calculate and print MSE for 'y1'
y1_mse = mean_squared_error(df['y1'], y1_pred)
print(f"Mean Squared Error (MSE) for y1: {y1_mse:.2f}")

# 7. Calculate and print R-squared for 'y1'
y1_r2 = r2_score(df['y1'], y1_pred)
print(f"R-squared for y1: {y1_r2:.2f}")

print("\n--- Error Metrics for y2 ---")
# 8. Calculate and print MSE for 'y2'
y2_mse = mean_squared_error(df['y2'], y2_pred)
print(f"Mean Squared Error (MSE) for y2: {y2_mse:.2f}")

# 9. Calculate and print R-squared for 'y2'
y2_r2 = r2_score(df['y2'], y2_pred)
print(f"R-squared for y2: {y2_r2:.2f}")
```

```python
import matplotlib.pyplot as plt

# Create a figure and two subplots
plt.figure(figsize=(12, 5))

# Subplot for y1 regression
plt.subplot(1, 2, 1) # 1 row, 2 columns, first plot
plt.scatter(df['x'], df['y1'], label='Original y1 data', color='blue')
plt.plot(df['x'], y1_pred, label='y1 regression line', color='red')
plt.title('y1 Regression')
plt.xlabel('x')
plt.ylabel('y1')
plt.legend()
plt.grid(True)

# Subplot for y2 regression
plt.subplot(1, 2, 2) # 1 row, 2 columns, second plot
plt.scatter(df['x'], df['y2'], label='Original y2 data', color='green')
plt.plot(df['x'], y2_pred, label='y2 regression line', color='orange')
plt.title('y2 Regression')
plt.xlabel('x')
plt.ylabel('y2')
plt.legend()
plt.grid(True)

plt.tight_layout() # Adjust layout to prevent overlapping titles/labels
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Re-create DataFrame 'df' (from original cell a14da8fb)
x = [1, 2, 3]
y1 = [2, 4, 6]
y2 = [3, 5, 7]
df = pd.DataFrame({'x': x, 'y1': y1, 'y2': y2})

# Re-create and fit models 'model_y1', 'model_y2' and 'X' (from original cell d8807eab)
X = df['x'].values.reshape(-1, 1)
model_y1 = LinearRegression()
model_y1.fit(X, df['y1'])
model_y2 = LinearRegression()
model_y2.fit(X, df['y2'])

# Make predictions for 'y1'
y1_pred = model_y1.predict(X)
# Make predictions for 'y2'
y2_pred = model_y2.predict(X)

# Calculate MSE and R-squared for 'y1'
y1_mse = mean_squared_error(df['y1'], y1_pred)
y1_r2 = r2_score(df['y1'], y1_pred)

# Calculate MSE and R-squared for 'y2'
y2_mse = mean_squared_error(df['y2'], y2_pred)
y2_r2 = r2_score(df['y2'], y2_pred)

# Create a DataFrame for error metrics
error_metrics = pd.DataFrame({
    'Metric': ['MSE', 'R-squared'],
    'y1_value': [y1_mse, y1_r2],
    'y2_value': [y2_mse, y2_r2]
})

# Reshape the DataFrame for seaborn plotting
error_metrics_melted = error_metrics.melt(id_vars='Metric', var_name='Target Variable', value_name='Value')
error_metrics_melted['Target Variable'] = error_metrics_melted['Target Variable'].str.replace('_value', '')

# Plot for y1
plt.figure(figsize=(7, 5))
sns.barplot(x='Metric', y='Value', data=error_metrics_melted[error_metrics_melted['Target Variable'] == 'y1'], palette='Blues_d')
plt.title('Error Metrics for y1 Model')
plt.ylabel('Value')
plt.xlabel('Metric')
plt.ylim(0, 1.1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Plot for y2
plt.figure(figsize=(7, 5))
sns.barplot(x='Metric', y='Value', data=error_metrics_melted[error_metrics_melted['Target Variable'] == 'y2'], palette='Greens_d')
plt.title('Error Metrics for y2 Model')
plt.ylabel('Value')
plt.xlabel('Metric')
plt.ylim(0, 1.1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
