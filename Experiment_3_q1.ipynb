```python
import numpy as np
from sklearn.linear_model import Ridge

# Define the dataset
X = np.array([2, 3, 4]).reshape(-1, 1) # Reshape X to be a 2D array
y = np.array([1, 2, 3])

# Initialize the Ridge regression model with alpha (lambda) = 1
ridge_model = Ridge(alpha=1)

# Fit the model to the data
ridge_model.fit(X, y)

# Get the regression coefficient
coefficient = ridge_model.coef_[0]

print(f"The regression coefficient for Ridge Regression (lambda=1) is: {coefficient:.4f}")
```
